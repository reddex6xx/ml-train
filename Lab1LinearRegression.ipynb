```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import mglearn as mg
# подключаем библиотеку для построения линейной регрессии
from sklearn.linear_model import LinearRegression
```

```python
#генерируем набор данных
X, y = mg.datasets.make_wave(n_samples=60)
#строим график
plt.scatter(X,y)
plt.xlabel("Признак")
plt.ylabel("Целевая переменная")
plt.show()
```

```python
#строим линейную регрессию
lr = LinearRegression()
lr.fit(X,y);
```

```python
#посчитаем коэффициент детерминации (R^2)
lr.score(X,y)
```

```python
#выводим коэффициенты b и k из уравнения y=kx+b; в лекциях b - w0, k - w1
lr.intercept_, lr.coef_
```

```python
#предсказываем значения y по уравнение регрессии
y_pred = lr.predict(X)
```

```python
#строим график
plt.scatter(X,y, color = '#ff00ff')
plt.xlabel("Признак")
plt.ylabel("Целевая переменная")
plt.plot(X, y_pred, color = '#00ff00')
plt.show()
```
